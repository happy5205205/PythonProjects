```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib notebook
# % matplotlib inline
```

```python
# 线性图
linear_data = np.arange(1, 9)
quadratic_data = linear_data ** 2
plt.figure()
plt.plot(linear_data,'-o', quadratic_data, '-o')
# 注意，这里我们只指定了y轴数据，x轴的数据是matplotlib自动生成的
```

```python
plt.plot([30, 40,50],'--r')
```

```python
plt.title('linear')
plt.xlabel('x data')
plt.ylabel('y data')
plt.legend(['legend1', 'legend2', 'legend3'], loc=2, frameon=False)
```

```python
plt.gca().fill_between(range(len(linear_data)),
                       linear_data, quadratic_data,
                       facecolor='green',
                       alpha=0.5
                      )
```

```python
# 绘制横轴为时间的线图
import numpy as np
plt.figure()
observation_dates = np.arange('2019-05-23', '2019-05-31', dtype='datetime64[D]')
# observation_dates
plt.plot(observation_dates, linear_data,'-o',  observation_dates, quadratic_data, '-o')
```

```python
# 横轴并不是我们想要的结果 对横坐标进行修改旋转45度
# 借助pandas绘制横轴为时间的线图
import pandas as pd
plt.figure()
observation_dates = np.arange('2019-05-23', '2019-05-31', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))
plt.plot(observation_dates, linear_data,'-o',  observation_dates, quadratic_data, '-o')
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
# 调整边界距离
plt.subplots_adjust(bottom=0.25)
```

```python
# 对于学术制图，可在标题中包含latex语法
ax = plt.gca()
ax.set_title('Quadratic ($x^2$) vs. Linear ($x$)')
```

```python
# 柱状图
plt.figure()
x_val = list(range(len(linear_data)))
plt.bar(x_val, linear_data, width=0.3)
```

```python
x_val
```

```python
# group bar chart
# 同一副图中添加新的柱状图
# 注意，为了不覆盖第一个柱状图，需要对x轴做偏移
x_val2 = [item + 0.3 for item in x_val]
plt.bar(x_val2, quadratic_data, width=0.3)
```

```python
# stack bar chart
plt.figure()
x_vals = list(range(len(linear_data)))
plt.bar(x_vals, linear_data, width=0.3)
plt.bar(x_vals, quadratic_data, width=0.3, bottom=linear_data)
```

```python
# 横向柱状图
plt.figure()
x_vals = list(range(len(linear_data)))
plt.barh(x_vals, linear_data, height=0.3)
plt.barh(x_vals, quadratic_data, height=0.3, left=linear_data)
```

```python
report_2015 = pd.read_csv('2015.csv')
```

```python
report_2015
```

```python
grouped = report_2015.groupby('Region')
```

```python
grouped['Country'].count()
```

```python
# 迭代groupby对象
for group, frame in grouped:
    print('{}地区的最大{:.3f}最小{:.3f}'.format(group, frame['Happiness Score'].max(),frame['Happiness Score'].min()))
```

```python
# 自定义函数进行分组
# 按照幸福指数排名进行划分，1-10, 10-20, >20
# 如果自定义函数，操作针对的是index
report_data2 = report_2015.set_index('Happiness Rank')
def get_rank_group(rank):
    rank_group = ''
    if rank <= 10:
        rank_group = '0--10'
    elif rank <= 20:
        rank_group = '10--20'
    else:
        rank_group = '大于20'
    return rank_group
groups= report_data2.groupby(get_rank_group)
for group, frame in groups:
#     print('{}最大值{}'.format(group,frame['Happiness Score'].max()))
    print('{}分组的数据个数{}'.format(group,len(frame)))
```

```python
# 实际项目中，通常可以先人为构造出一个分组列，然后再进行groupby

# 按照score的整数部分进行分组
# 按照幸福指数排名进行划分，1-10, 10-20, >20
# 如果自定义函数，操作针对的是index
report_data2['score'] = report_data2['Happiness Score'].apply(lambda x:int(x))
groups = report_data2.groupby('score')
for group, frame in groups:
    print('幸福指数整数位{}的个数{}'.format(group, len(frame)))
```
